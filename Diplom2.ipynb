```python
from bs4 import BeautifulSoup

# указываем путь к файлу .plx
filename = 'plan.plx'

# читаем содержимое файла
with open(filename, 'r', encoding='utf-16') as file:
    content = file.read()

# парсим содержимое файла с помощью BeautifulSoup
soup = BeautifulSoup(content, 'xml')

# ищем все элементы ПланыСтроки и ПланыНовыеЧасы
plan_rows = soup.find_all('ПланыСтроки')
plan_hours = soup.find_all('ПланыНовыеЧасы')


# Дисциплины
diciplines_names_dict = {}
discipline_names = soup.find_all('ПланыСтроки')
for discip in discipline_names:
    diciplines_names_dict[discip.get('Код')] = discip.get('Дисциплина')

# Кафедры
department_dict = {}
departments = soup.find_all('Кафедры')
for dept in departments:
    department_dict[dept.get('Код')] = dept.get('Название')

# СправочникВидыРабот 
type_work_dict = {}
type_work = soup.find_all('СправочникВидыРабот')
for works in type_work:
    type_work_dict[works.get('Код')] = works.get('Название')
    
# СправочникВидОбъекта
view_object_dict = {}
view_object = soup.find_all('СправочникВидОбъекта')
for view in view_object:
    view_object_dict[view.get('Код')] = view.get('Наименование')
    
# СправочникВидыПрактик
type_practice_dict = {}
type_practice = soup.find_all('СправочникВидыПрактик')
for practice in type_practice:
    type_practice_dict[practice.get('Код')] = practice.get('Наименование')
    
# СправочникТипаЧасов
type_hours_dict = {}
type_hours = soup.find_all('СправочникТипаЧасов')
for hours in type_hours:
    type_hours_dict[hours.get('Код')] = hours.get('Наименование')

# СправочникТипБлока
types_block_dict = {}
types_block = soup.find_all('СправочникТипБлока')
for block in types_block:
    types_block_dict[block.get('Код')] = block.get('Название')
    
# СправочникТипОбъекта
types_object_dict = {}
types_object = soup.find_all('СправочникТипОбъекта')
for objects in types_object:
    types_object_dict[objects.get('Код')] = objects.get('Название')

# ФормаОбучения
form_education_dict = {}
form_education = soup.find_all('ФормаОбучения')
for forms in form_education:
    form_education_dict[forms.get('Код')] = forms.get('ФормаОбучения')

# СправочникВидыРабот

    
# проходим по всем элементам и выводим на экран Дисциплину, ДисциплинаКод и Кафедру
for row in plan_rows:
    for plans in plan_hours:
        discipline = row.get('Дисциплина')
        discipline_code = row.get('ДисциплинаКод')
        department_code = row.get('КодКафедры')
        department_name = department_dict.get(department_code)
        type_hours_code = plans.get('КодТипаЧасов')
        type_hours_name = type_hours_dict.get(type_hours_code)
        type_work_code = plans.get('КодВидаРаботы')
        type_work_name = type_work_dict.get(type_work_code)
        course = plans.get('Курс')
        semestr = plans.get('Семестр')
        hours_value = plans.get('Количество')
        print(f'Дисциплина: {discipline},\n Код дисциплины: {discipline_code},\n Кафедра: {department_name}, \n Тип работ: {type_work_name}, \n Тип Часов: {type_hours_name}, \n Количесвто часов: {hours_value}')
```

```python
import csv

# определяем имена столбцов
headers = ['Номер курса', 'Семестр', 'Название дисциплины', 'Кафедра', 'шифры компетенций', 'Тип работы', 'Тип часов', 'Часы']

# создаем csv-файл и записываем заголовок
with open('data_rpid.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(headers)

    # перебираем все дисциплины и записываем каждую строку в файл csv
    for code, data in data_by_discipline.items():
        discipline_name = data['name']
        discipline_code = code
        department_name = data['department']

        # получаем список всех компетенций и их шифры для данной дисциплины
        competencies = soup.find_all('ПланыКомпетенцииДисциплины', {'КодСтроки': code})
        competency_codes = [comp['КодКомпетенции'] for comp in competencies]
        competency_names = [soup.find('ПланыКомпетенции', {'Код': comp})['ШифрКомпетенции'] for comp in competency_codes]
        competency_codes_str = ','.join(competency_names)

        # перебираем все часы для данной дисциплины и записываем каждую строку в файл csv
        for hour_data in data['hours']:
            work_type = hour_data['work_type']
            hour_type = hour_data['hour_type']
            hours = hour_data['hours']
            course = hour_data['course']

            # определяем семестр на основе курса и семестра
            if int(course) == 1:
                semester = int(hour_data['semester'])
                if semester == 1:
                    semester = 1
                elif semester == 2:
                    semester = 2
            elif int(course) == 2:
                semester = int(hour_data['semester']) + 2
            elif int(course) == 3:
                semester = int(hour_data['semester']) + 4
            elif int(course) == 4:
                semester = int(hour_data['semester']) + 6
            elif int(course) == 5:
                semester = int(hour_data['semester']) + 8

            # записываем строку в файл csv
            row = [course, semester, discipline_name, department_name, competency_codes_str, work_type, hour_type, hours]
            writer.writerow(row)
```
